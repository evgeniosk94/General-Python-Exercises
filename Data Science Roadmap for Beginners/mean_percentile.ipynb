```python
percentile_97 = df.price.quantile(0.97)
df_97 = df[df.price >= percentile_97]
df_97.shape
```

```python
percentile_0 = df.price.quantile(0.03)
df_0 = df[df.price <= percentile_0]
df_0
```

```python
df_no_outlier = df[(df.price <= percentile_97) & (df.price >= percentile_0)]
df_no_outlier
```
