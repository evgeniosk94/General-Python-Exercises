```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("bhp.csv")
df
```

```python
df.describe()
```

```python
percentile_upper = df.total_sqft.quantile(0.999)
percentile_upper
```

```python
percentile_lower = df.total_sqft.quantile(0.001)
percentile_lower
```

```python
df_new_percentiles = df[(df.total_sqft >= percentile_lower) & (df.total_sqft <= percentile_upper)]
df_new_percentiles
```

```python
df_new_percentiles.describe()
```

```python
standard_dev = df.total_sqft.std()
standard_dev
mean = df.total_sqft.mean()
```

```python
df_new_std = df[(df.total_sqft >= mean - 4*standard_dev) & (df.total_sqft <= mean + 4*standard_dev)]
df_new_std
```

```python
sns.histplot(df_new_std, x="total_sqft", kde=True)
```

```python
df["zscore"] = (df.total_sqft - df.total_sqft.mean()) / df.total_sqft.std()
df.head()
```

```python
df_new_z_score = df[(df.zscore >= -4) & (df.zscore <= 4)]
df_new_z_score
```
