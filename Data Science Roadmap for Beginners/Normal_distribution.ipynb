```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("bhp.csv")
df
```

```python
df.describe()
```

```python
percentile_upper = df.total_sqft.quantile(0.999)
percentile_upper
```

```python
percentile_lower = df.total_sqft.quantile(0.001)
percentile_lower
```

```python
df_new_percentiles = df[(df.total_sqft >= percentile_lower) & (df.total_sqft <= percentile_upper)]
df_new_percentiles
```

```python
df_new_percentiles.describe()
```

```python
standard_dev = df.total_sqft.std()
standard_dev
mean = df.total_sqft.mean()
```

```python
df_new_std = df[(df.total_sqft >= mean - 4*standard_dev) & (df.total_sqft <= mean + 4*standard_dev)]
df_new_std
```

```python
sns.histplot(df_new_std, x="total_sqft", kde=True)
```

```python
df["zscore"] = (df.total_sqft - df.total_sqft.mean()) / df.total_sqft.std()
df.head()
```

```python
df_new_z_score = df[(df.zscore >= -3) & (df.zscore <= 3)]
df_new_z_score
```

```python
_, mean, std, _, _, median, *_ = df.total_sqft.describe()    # _ t ostore a variable that you dont care, *_ to do the same for all the rest
```

```python
MAD = np.median((np.abs(df.total_sqft -median)))
def get_zscore_modified(value, median, MAD):
        z_score_modified = 0.6745 * (value - median) / MAD
        return z_score_modified
```

```python
df["z_score_modified"] = df.total_sqft.apply(lambda x: get_zscore_modified(x, median, MAD))
df.head()
```

```python
df[df.zscore>4]
```

```python
df_new_z_score_mod=df[df.z_score_modified<3.5]
```
