```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
df = pd.DataFrame(digits.data, columns=digits.feature_names)
df
```

```python
df["digits"] = digits.target
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
```

```python
cross_val_score(SVC(kernel="rbf", C=10, gamma="auto"), digits.data, digits.target, cv=5)
```

```python
cross_val_score(SVC(kernel="linear", C=10, gamma="auto"), digits.data, digits.target, cv=5)
```

```python
cross_val_score(SVC(kernel="linear", C=20, gamma="auto"), digits.data, digits.target, cv=5)
```

```python
kernels = ["rbf", "linear"]
C = [1,10,20]
avg_scores = {}
for kval in kernels:
        for cval in C:
                cv_scores = cross_val_score(SVC(kernel=kval, C=cval, gamma="auto"), digits.data, digits.target, cv=5)
                avg_scores[kval + '_' + str(cval)] = np.average(cv_scores)
avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma="auto"), {
        "C": [1,10,20],
        "kernel": ["rbf", "linear"]
}, cv=5, return_train_score=False)

clf.fit(digits.data, digits.target)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[["param_C", "param_kernel", "mean_test_score"]]
```

```python
clf.best_params_
```

```python
model_params = {
    'svm': {
        'model': SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20,50],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10,20,50,70,100,150]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear'),
        'params': {
            'C': [1,5,10,20]
        }
    },
    "decision_trees": {
        "model": DecisionTreeClassifier(),
        "params": {
             "max_depth": [1,3,5,10,20,30],
             "criterion": ["gini", "entropy"]    
        }
    }
}
```

```python
scores = []

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=cv, return_train_score=False)
    clf.fit(digits.data, digits.target)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
